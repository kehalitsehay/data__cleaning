```python
import pandas as pd
```

```python
df = pd.read_excel('Customer Call List.xlsx')
df         
```

```python
df.info()
```

```python
df = df.drop_duplicates()
```

```python
df
```

As you can see, in the last_name columns there are additional unwanted characters such as (/.._). 

```python
df['Last_Name']=df['Last_Name'].str.strip('.._/')
df
```

The phone number columns should be standardized and have a similar format for a better layout. To do so, we first remove all characters and present the numbers.

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('[^0-9]', '', regex=True)
df
```

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x))
df
```

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6]  + '-' + x[6:10])
df
```

```python
# Now most of the phone number is in the standard way. the only thing is removing the (nan--,--)
df['Phone_Number'] = df['Phone_Number'].str.replace('nan--', '')
df['Phone_Number'] = df['Phone_Number'].str.replace('--', '')
df
```

```python
# The next coulumn  that we could work is about the address column
df[['Street_Address', 'State', 'Zip']] = df['Address'].str.split(',', n=2, expand=True)
df
```

```python
df = df.drop(columns="Address")
```

```python
df
```

```python
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes', 'Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No', 'N')
df['Do_Not_Contact'] = df['Do_Not_Contact'] .str.replace('Yes', 'Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No', 'N')
```

```python
df
```

```python
# To have a smooth relationship with the customer, let's remove those customers who don't want to contact
for x in df.index:
    if df.loc[x, 'Do_Not_Contact'] == 'Y':
        df.drop(x, inplace = True)
df
```

```python
# In addtion, lets remove customers who's phone number is not registred
for x in df.index:
    if df.loc[x, 'Phone_Number'] == "":
        df.drop(x, inplace=True)

df
```

```python
# Now let's reset the index for better expreince 
df.reset_index(inplace=True, drop=True)
df
```

```python
df = df.drop(columns='index')
```
